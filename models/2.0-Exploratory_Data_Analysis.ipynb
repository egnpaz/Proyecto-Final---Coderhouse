```python
# Importing data from the clean .csv data file.

df = pd.read_csv ('heart_disease_clean.csv', index_col = 0)
```

```python
#We can see that the variable with more deviation is our target variable. Let do a Univariate analysis to explore this
#a little more.
```

# 
# Exploring numerical variables
# 

```python
# AGE
#Age Groups:

#1  18-24 years
#2  25-29 years
#3  30-34 years
#4  35-39 years
#5  40-44 years
#6  45-49 years
#7  50-54 years
#8  55-59 years
#9  60-64 years
#10 65-69 years
#11 70-74 years
#12 75-79 years
#13 80 or older years

fig, (ax1, ax2, ax3) = plt.subplots (1,3, figsize = (15,4))
plt.style.use ('fivethirtyeight')
ax1. set_title ('Age in 5 years groups')
ax1.set_ylabel ('Years')
ax1.boxplot (x= 'Age', data= df, showmeans = True)

ax2.set_title ('Age in 5 years groups')
ax2.set_xlabel ('Years')
ax2.hist (x= 'Age', data= df, edgecolor = 'black')

ax3.set_title ('Age in 5 years groups')
stats.probplot (df['Age'], dist = 'norm', plot = plt)
plt.show ()

estimate, p_value= shapiro (df['Age'])
print ('Estimate=%.3f, p_value=%.3f' % (estimate, p_value))
```

```python
# We see the age the age has aproximately normal distribution ,in the boxplot show the median in a symmetric position in
# relation with the interquantile range, the whiskers are also al most symmetric and there are no outliers, the histogram shows
# also an aproximately gaussian distribution with a little skew to the right. The QQ plot shows that almost all the data aligns
# with the probability line.
# The mean age of our population is around group 8 wich is 55-59 years, an age where heart diseases are more frequent.The median
#is a little higher around 9.
```

```python
df.dtypes
```

```python
# WEIGHT


fig, (ax1, ax2, ax3) = plt.subplots (1,3, figsize = (15,4))
plt.style.use ('fivethirtyeight')
ax1. set_title ('Weight in Kg')
ax1.set_ylabel ('Kilograms')
ax1.boxplot (x= 'Weight', data= df, showmeans = True)

ax2.set_title ('Weight in Kg')
ax2.set_xlabel ('Kilograms')
ax2.hist (x= 'Weight', data= df, edgecolor = 'black')

ax3.set_title ('Weight in Kg')
stats.probplot (df['Weight'], dist = 'norm', plot = plt)
plt.show ()

estimate, p_value= shapiro (df['Weight'])
print ('Estimate=%.3f, p_value=%.3f' % (estimate, p_value))
```

```python
#The boxplot is symmetric and th mean and median are similar (around 175 kg.) wich talks of an aproximately normal
#distribution. The are several outliers but these extreme values (obese people) are related with the disease. The histogram
#has also an normal distribution a little skewed to the left. The probability plot shows a deviation from the normal line
#in the extremes related to the outliers.
```

```python
#BMI

fig, (ax1, ax2, ax3) = plt.subplots (1,3, figsize = (15,4))
plt.style.use ('fivethirtyeight')
ax1. set_title ('Body Mass Index')
ax1.set_ylabel ('BMI Units')
ax1.boxplot (x= 'BMI', data= df, showmeans = True)

ax2.set_title ('Body Mass Index')
ax2.set_xlabel ('BMI Units')
ax2.hist (x= 'BMI', data= df, edgecolor = 'black')

ax3.set_title ('Body Mass Index')
stats.probplot (df['Weight'], dist = 'norm', plot = plt)
plt.show ()

estimate, p_value= shapiro (df['BMI'])
print ('Estimate=%.3f, p_value=%.3f' % (estimate, p_value))
```

```python
#Since the BMI is related to the wieght we see a similar pattern here. The outliers don't need to be corrected since they
#are related to obese people.
```

# 2. Bivariate Analysis
# -----------------------------------------------------------------------------------------------------------

```python
#We are going to asses the correlation between all the variables
df.corr ()
```

```python
# We can see a positive association with risk factors like cholesterol and high blood pressure anda a negative association
#for benefitial factor like eating fruits anda vegetables.
```

```python
# In general, risk factors for heart disease tend to occur more frequently in men than in young women
# due to a hormonal factor, they present protection against these. In postmenopausal women these effects disappear and
#the risk of disease is equal to that of men.

#For this reason, it is interesting to evaluate the average age of our population according to sex.

df.groupby ('Sex').Age.mean ()
```

```python
# We can see that both sexes have a high and similar average age (group 8 55-59 years). We can also evaluate this graphically with a barplot

plt.figure (figsize = (6, 6))
plt.title ('Relacion Edad segun Sexo')
sns.barplot (x = 'Sex', y = 'Age', data = df, palette = 'hot')
plt.ylabel ('Edad')
plt.xlabel ('Sexo, False= Mujer True = Hombre')
```

```python
#We can graphically evaluate how different risk factors are distributed according to sex.

plt.figure (figsize = (6,6))
plt.title ('Weight by Sex')
sns.violinplot ( x = 'Sex', y = 'Weight', data = df, inner = 'quartile')
plt.ylabel ('Weight in Kilograms')
plt.xlabel ('Sex, Flase= Female True= Male')
```

```python
plt.figure (figsize = (6,6))
plt.title ('BMI by Sex')
sns.violinplot ( x = 'Sex', y = 'BMI', data = df, inner = 'quartile')
plt.ylabel ('BMI Units')
plt.xlabel ('Sex, Flase= Female True= Male')
```

```python
plt.style.use ('fivethirtyeight')
plt.title ('High Blood Pressure by Sex')
plt.xlabel ('High Blood Pressure')
sns.countplot(data=df, x="High_BP", hue="Sex")
```

```python
plt.style.use ('fivethirtyeight')
plt.title ('Diabetes by Sex')
plt.xlabel ('Diabetes')
sns.countplot(data=df, x="Diabetes", hue="Sex")
```

```python
# Through the violin graphs we can see that there is no great difference in the distribution of risk factors between both sexes
#The countplots show a slight prevalence of risk factores in women (no to be expected) but this is probably due to the imba
#lance between both sexes in the population.
```

# Multivariate Analysis

```python
#Exploring the correlation of multiple variables trough a pairplot graph.

plt.style.use ('seaborn')
sns.pairplot (df, palette = 'viridis')
```

```python
#Since most of the variables are categorical is difficult to analyze the pairplot. We can see a strong correlation between
#Weight and BMI wich is expected since BMI is calculates using the weight values.


#We can also use a heatmat to graphically evaluate the relationship between our variables.

plt.figure (figsize = (15,15))
sns.heatmap (df.corr (), vmin = -1, vmax = 1, linewidth = 0.4, annot = True)
plt.show ()
```

```python
#In our heatmap show that our target variable (hd (Heart Disease)) is more related to High Blood Pressure, 
#High Cholestrol, Had a previous Stroke and Diabeetes. While obesity a factor classically associated with heart disease 
#does not show such a strong relationship. 

# We can graphically analyze the association of various risk factors for heart disease.
```

```python
#We are going to graph the relationship between multiple risk factors, gender and then separate them by the target variable

plt.figure (figsize = (15,15))
g = sns.relplot (x = 'Age', y = 'Weight', data = df, hue = 'Sex', palette = ['purple', '#55CCCC'], style = 'Sex', s = 150, alpha = 0.8, col = 'Heart_Disease', kind = 'scatter')

g.set_xlabels ('Age in 5 Years Group')
g.set_ylabels ('Weight in Kg.')
plt.show ()
```

```python
# We can see in this graph that older people and the ones who had Heart Disease tend to have a higher weight, it doesn't
#sem to be difference between both sexes.
```
